### Exercise 1: Prime Numbers

```python
def is_prime(n):
    if n <= 1:
        return False
    elif n <= 3:
        return True
    elif n % 2 == 0 or n % 3 == 0:
        return False
    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6
    return True
```

```python
number = int(input("Enter a number to check if it's prime: "))
if is_prime(number):
    print(number, "is a prime number")
else:
    print(number, "is not a prime number")
```

### Exercise 2: Product of Random Numbers

```python
import random

# Generate two random integers between 1 and 10
num1 = random.randint(1, 10)
num2 = random.randint(1, 10)

# Ask the user to calculate the product
print(f"What is the product of {num1} and {num2}?")
user_answer = int(input("Your answer: "))

# Check the answer
correct_answer = num1 * num2

if user_answer == correct_answer:
    print(" Correct! Well done.")
else:
    print(f" Incorrect. The correct answer is {correct_answer}.")
```

### Exercise 3: Squares of Even/Odd Numbers

```python
# Python script to print the squares of all EVEN numbers from 100 to 200
print("Squares of even numbers from 100 to 200:\n")

for num in range(100, 201):  # range is inclusive of 100, exclusive of 201
    if num % 2 == 0:
        square = num ** 2
        print(f"{num}^2 = {square}")
```

### Exercise 4: Word counter

```python
# Take input from user
text = input("Enter your text: ")

# Split the text into words
words = text.split()

# Create a dictionary to store word counts
word_counts = {}

# Count the frequency of each word
for word in words:
    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

# Display the result
print("\nWord Frequencies:")
for word, count in word_counts.items():
    print(f"'{word}': {count}")
```

### Exercise 5: Check for Palindrome

```python
def is_palindrome(s):
    # Convert the string to lowercase and remove spaces and punctuation
    s = ''.join(char.lower() for char in s if char.isalnum())

    # Check if the string is equal to its reverse
    return s == s[::-1]

# Example usage
input_str = "racecar"
print(is_palindrome(input_str))  # Output: True
```
