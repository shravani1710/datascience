# CHI-SQUARE TEST

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
# Contingency table
data = np.array([[50, 70],
                     [80, 100],
                     [60, 90],
                     [30, 50],
                     [20, 50]])
```

### 1. State the Hypotheses

##### Null Hypothesis (H0): Device type and customer satisfaction are independent.
##### Alternative Hypothesis (H1): There is an association between device type and customer satisfaction.

#### 2. Compute the Chi-Square Statistic:

```python
# Perform chi-square test for independence
chi2, p, dof, expected = chi2_contingency(data)
```

```python
# Print results
print("Chi-Square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("p-value:", p)
```

#### 3. Determine the Critical Value:

```python
from scipy.stats import chi2
alpha = 0.05
dof = 4
critical_value = chi2.ppf(1 - alpha, dof)  # Degrees of freedom = (rows - 1) * (columns - 1)
print("Critical Value:", critical_value)
```

#### 4. Make a Decision:

```python
if chi2 > critical_value:
    print("Reject the null hypothesis.")
    print("Conclusion: There is a significant association between device type and customer satisfaction.")
else:
    print("Fail to reject the null hypothesis.")
    print("Conclusion: No significant association found between device type and customer satisfaction.")
```
